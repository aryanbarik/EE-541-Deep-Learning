uniform random samples whose sum exceeds four. Generate a histogram using 100, 1000, and

```python
import random
import matplotlib.pyplot as plt
```

```python
def get_n():
    x = 0
    n = 0
    
    while(x < 4):
        n += 1
        x += random.uniform(0,1)
        # print(x)
    
    return n
```

```python
def get_n_x_times(x):
    n_appearances = []
    
    for i in range(x):
        n_appearances.append(get_n())
    
    plt.hist(n_appearances)        
```

```python
get_n_x_times(100)
```

```python
get_n_x_times(1000)
```

```python
get_n_x_times(10000)
```
