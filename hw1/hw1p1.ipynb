## 1a

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
flips = []

for i in range(50):
    flips.append(random.random())

print(flips)

total_heads = 0
longest_run_heads = 0
current_run_heads = 0
is_last_flip_heads = False

# dont lose greatest value from streak to streak

for i in flips:
    if i >= 0.30:
        total_heads += 1
        if is_last_flip_heads:
            current_run_heads += 1
        
        else:    
            is_last_flip_heads = True
            current_run_heads = 1

    else:
        is_last_flip_heads = False
        current_run_heads = 0
    
    if current_run_heads > longest_run_heads:
        longest_run_heads = current_run_heads

print("Longest run of heads is:", longest_run_heads)
print("Total number of heads is:", total_heads)
```

## 1b

```python
def fifty_flip_experiment():
    flips = []

    for i in range(50):
        flips.append(random.random())

    total_heads = 0

    for i in flips:
        if i >= 0.30:
            total_heads += 1
            
    return total_heads
```

```python
def repeat_experiment(x):
    heads = []
    for i in range(x):
        heads.append(fifty_flip_experiment())

    print(heads)

    # histogram
    plt.hist(heads)    
```

```python
repeat_experiment(20)
```

```python
repeat_experiment(100)
```

```python
repeat_experiment(200)
```

```python
repeat_experiment(1000)
```

One of the limits of the histogram is that the displayed information can be manipulated by changing the size of the bins. This could leave a viewer with a skewed impression of the data.
